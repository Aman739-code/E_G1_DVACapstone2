# 03 - Exploratory Data Analysis (EDA)

In this notebook, we explore the cleaned Amazon product data to uncover initial insights.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure plots appear inline
%matplotlib inline
```

### Load Cleaned Data

```python
df = pd.read_csv('../data/processed/cleaned_data.csv')
df.info()
```

### Summary Statistics

```python
df.describe()
```

### Price Distribution

```python
plt.figure(figsize=(10, 6))
plt.hist(df['current_price'], bins=50, range=(0, 200), color='skyblue', edgecolor='black')
plt.title('Distribution of Current Prices (up to $200)')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

### Top 10 Reviewed Products

```python
top_reviewed = df.sort_values(by='number_of_reviews', ascending=False).head(10)
top_reviewed[['title', 'rating', 'number_of_reviews', 'current_price']]
```
