```python
import pandas as pd

DATA_PATH = '../data/processed/cleaned_data.csv'
TABLEAU_READY_PATH = '../data/processed/tableau_ready_dataset.csv'

df = pd.read_csv(DATA_PATH)
df.head()
```

### KPI Calculations

We compute the following metrics for the dataset:
- **Avg Rating**: `AVG([Rating])`
- **Discount %**: `([Listed Price] - [Current Price]) / [Listed Price] * 100`
- **Best Seller Rate %**: `SUM(IIF([Is Best Seller] = "1", 1, 0)) / COUNT([Title]) * 100`
- **Coupon Adoption %**: `SUM(IIF([Has Coupon] = "1", 1, 0)) / COUNT([Title]) * 100`
- **Avg Monthly Sales**: `AVG([Bought In Last Month])`



```python
# Row-level calculations for Tableau
df['discount_percent'] = ((df['listed_price'] - df['current_price']) / df['listed_price']) * 100

# Aggregated KPI Summary for Verification
kpi_summary = {
    'Avg Rating': df['rating'].mean(),
    'Discount %': df['discount_percent'].mean(), 
    'Best Seller Rate %': (df['is_best_seller'] == 1).sum() / df['title'].count() * 100,
    'Coupon Adoption %': (df['has_coupon'] == 1).sum() / df['title'].count() * 100,
    'Avg Monthly Sales': df['bought_in_last_month'].mean()
}

kpi_df = pd.DataFrame(list(kpi_summary.items()), columns=['KPI', 'Value'])
kpi_df
```

```python
df.to_csv(TABLEAU_READY_PATH, index=False)
print(f'Saved Tableau-ready dataset to {TABLEAU_READY_PATH}')
```
